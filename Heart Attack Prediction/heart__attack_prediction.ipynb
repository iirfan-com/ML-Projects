```python
#encoding
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
# Assuming 'column_name' is the categorical column you want to encode
df['Gender'] = encoder.fit_transform(df['Gender'])
df['Smoking_Status'] = encoder.fit_transform(df['Smoking_Status'])
# Assuming 'column_name' is the categorical column you want to encode
df['Alcohol_Intake'] = encoder.fit_transform(df['Alcohol_Intake'])
# Assuming 'column_name' is the categorical column you want to encode
df['Physical_Activity'] = encoder.fit_transform(df['Physical_Activity'])
# Assuming 'column_name' is the categorical column you want to encode
df['Diabetes_Status'] = encoder.fit_transform(df['Diabetes_Status'])
# Assuming 'column_name' is the categorical column you want to encode
df['Family_History_Heart_Disease'] = encoder.fit_transform(df['Family_History_Heart_Disease'])
# Assuming 'column_name' is the categorical column you want to encode
df['Diet_Quality'] = encoder.fit_transform(df['Diet_Quality'])
# Assuming 'column_name' is the categorical column you want to encode
df['Stress_Level'] = encoder.fit_transform(df['Stress_Level'])
# Assuming 'column_name' is the categorical column you want to encode
df['Heart_Attack_History'] = encoder.fit_transform(df['Heart_Attack_History'])
# Assuming 'column_name' is the categorical column you want to encode
df['Medication_Usage'] = encoder.fit_transform(df['Medication_Usage'])




```

```python
df.dtypes
```

```python
df.drop(["Patient_ID"],axis=1,inplace=True)
df
```

```python
x=df.drop(["Heart_Attack_Outcome"],axis=1)
y=df["Heart_Attack_Outcome"]
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
x_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=7)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Make predictions on the test data
y_pred =model.predict(x_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Calculate precision
precision = precision_score(y_test, y_pred)
print("Precision:", precision)

# Calculate recall
recall = recall_score(y_test, y_pred)
print("Recall:", recall)

# Calculate F1-score
f1 = f1_score(y_test, y_pred)
print("F1-score:", f1)
```
