```python
correlation = df.corr()["Bankrupt?"].abs().sort_values(ascending=False)

# Display the top 10 most correlated features
print(correlation[1:6])

# Plot correlation heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df[correlation.index[:6]].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Top Features Correlated with Bankruptcy")
plt.show()
```

```python
roa_by_bankruptcy = df.groupby('Bankrupt?')[' ROA(C) before interest and depreciation before tax'].mean()

# Create a bar chart
plt.figure(figsize=(8, 6))
sns.barplot(x=roa_by_bankruptcy.index, y=roa_by_bankruptcy.values,color='red',width=0.4)
plt.title('Average ROA before tax for Bankrupt vs. Non-Bankrupt Companies')
plt.xlabel('Bankruptcy Status (0 = Non-Bankrupt, 1 = Bankrupt)')
plt.ylabel('Average ROA')
plt.xticks([0, 1], ['Non-Bankrupt', 'Bankrupt'])  # Set custom x-axis labels
plt.show()
```

```python
x= df.iloc[:,1:].values
y=df.iloc[:,0].values
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python

from sklearn.ensemble import RandomForestClassifier
random=RandomForestClassifier(n_estimators=100)
random.fit(x_train,y_train)
y_pred=random.predict(x_test)
y_pred
```

```python
from sklearn.metrics import recall_score,f1_score,confusion_matrix,classification_report,accuracy_score,ConfusionMatrixDisplay
ac=accuracy_score(y_test,y_pred)
ac
```

```python
confusion=confusion_matrix(y_test,y_pred)
confusion
```

```python
classification=classification_report(y_test,y_pred)
classification
```
